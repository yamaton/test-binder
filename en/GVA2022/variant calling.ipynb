Single Nucleaotide Variant calling

### Reference genome

Use E.Coli data downloaded in "Read mapping" section.

#### Index FASTA

This will create an index file with `.fai` suffix.

```bash
ref='../read_mapping/ncbi_dataset/data/GCF_000017985.1/GCF_000017985.1_ASM1798v1_genomic.fna'
samtools faidx "$ref"
```

#### Prepare sorted indexed BAM 

```bash
relpath="../read_mapping/bowtie2"

samtools view "$relpath"/SRR030257.sam -b -o "$relpath"/SRR030257.bam --threads 4 
samtools sort "$relpath"/SRR030257.bam -o "$relpath"/SRR030257.sorted.bam --threads 4
samtools index "$relpath"/SRR030257.sorted.bam -@ 4
```

#### Call genome variants

```bash
mkdir -p outputs
bam="../read_mapping/bowtie2/SRR030257.sorted.bam"
ref='../read_mapping/ncbi_dataset/data/GCF_000017985.1/GCF_000017985.1_ASM1798v1_genomic.fna'
bcf="outputs/SRR030257.bcf"
```

```bash
bcftools mpileup --fasta-ref "$ref" --output-type 'u' --output "$bcf" --threads 8 "$bam"
```

#### BCF $\to$ VCF

```bash
bcf="outputs/SRR030257.bcf"
vcf="outputs/SRR030257.vcf"
bcftools call --variants-only --consensus-caller --threads 4 "$bcf" > "$vcf"
```

```bash
bcf="outputs/SRR030257.bcf"
bcftools stats "$bcf" --threads 2 "$bcf"
```

```bash
vcf="outputs/SRR030257.vcf"
awk -F ";" '{for(i=1;i<=NF;i++){if ($i ~ /AF1/){print $i}}}'  "$vcf"
```

#### Indexing BCF

```bash
bcf="outputs/SRR030257.bcf"
bcftools index "$bcf"
```

#### Indexing VCF

```bash
vcf="outputs/SRR030257.vcf"
bcftools index "$vcf"
```
