## All about the pangolin
https://www.biostarhandbook.com/books/corona/pangolin.html

```bash
bio search NC_045512
```

```bash
bio fetch NC_045512 MT121216 > genomes.gb
```

```bash
cat genomes.gb | bio fasta -gene E | bio align | head -5
```

```bash
cat genomes.gb  | bio fasta -gene E --translate | bio align | head -5
```

```bash
cat genomes.gb | bio fasta | mafft --auto - > aligned.fa
```

```bash
cat aligned.fa | bio format | head -12
```

```bash
cat genomes.gb
| bio fasta -gene S --translate | bio align --table
```

```bash
# Fetch data.
mkdir -p refs
# bio fetch NC_045512 --format fasta > refs/NC_045512.fa

bio fetch MT121216  --format fasta > refs/MT121216.fa
```

```bash
# Here we take the pangolin as the reference and align SARS2 to it:

minimap2 -a refs/MT121216.fa refs/NC_045512.fa \
  | samtools sort > aligned-pango-sars2.bam 
samtools index aligned-pango-sars2.bam
```

```bash
minimap2 -a  refs/NC_045512.fa refs/MT121216.fa \
  | samtools sort > aligned-sars2-pango.bam
samtools index aligned-sars2-pango.bam 
```
